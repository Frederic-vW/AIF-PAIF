## Examples

### [1] Markov chain of order 3

In this example, $X_{n+1}$ depends on $X_{n}, X_{n-1}, X_{n-2}$.

```python
x = np.load("./data/mc_1_3.npy").astype('int')
print(x.shape, x.dtype, set(x))
ns = len(set(x))
print(f"[+] Symbolic time series ns = {ns:d}")
plot_aif_paif(x, ns, kmax=10)
```

### [2] Another Markov chain or order 3

In this example, $X_{n+1}$ depends on $X_{n-2}$ only.

```python
x = np.load("./data/mc_3.npy").astype('int')
print(x.shape, x.dtype, set(x))
ns = len(set(x))
print(f"[+] Symbolic time series ns = {ns:d}")
plot_aif_paif(x, ns, kmax=10)
```

### [3] The non-Markovian even process

The even process is non-Markovian and in theory, has infinite memory.

```python
x = np.load("./data/even.npy").astype('int')
print(x.shape, x.dtype, set(x))
ns = len(set(x))
print(f"[+] Symbolic time series ns = {ns:d}")
plot_aif_paif(x, ns, kmax=10)
```

### [4] The 2-state golden-mean process

This Hidden Markov Model (HMM) representation of the golden mean process uses two hidden and two observable states. The Markov order is 1.

```python
x = np.load("./data/gm2.npy").astype('int')
print(x.shape, x.dtype, set(x))
ns = len(set(x))
print(f"[+] Symbolic time series ns = {ns:d}")
plot_aif_paif(x, ns, kmax=10)
```

### [5] The 7-state golden mean process

This alternative Hidden Markov Model implementation of the golden-mean process uses a Markov order of 4 and so-called cryptic order 3.

```python
x = np.load("./data/gm7.npy").astype('int')
print(x.shape, x.dtype, set(x))
ns = len(set(x))
print(f"[+] Symbolic time series ns = {ns:d}")
plot_aif_paif(x, ns, kmax=10)
```

### [6] The Ising model

The Ising model on a square finite lattice with periodic BC is run close to the critical point.

```python
x = np.load("./data/Ising2D_L50_N1000000_Temp2.27.npy").astype('int')
print(x.shape, x.dtype, set(x))
ns = len(set(x))
print(f"[+] Symbolic time series ns = {ns:d}")
plot_aif_paif(x, ns, kmax=10)
```

### [8] EEG microstate sequence

This example contains a 4-state symbolic sequence representing the sequence of spatial patterns measured by surface electroencephalography (EEG) from a healthy subject in a task-free (resting state) condition.

```python
x = np.load("./data/eeg_ms.npy").astype('int')
print(x.shape, x.dtype, set(x))
ns = len(set(x))
print(f"[+] Symbolic time series ns = {ns:d}")
plot_aif_paif(x, ns, kmax=10)
```
